```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import json

grey = np.ones((16,16),np.uint8);

for i in range(256):
    # print(i%16,i//16)
    grey[i%16][i//16] = i

rgb = cv.applyColorMap(grey,cv.COLORMAP_TURBO)
plt.imshow(rgb)

# print(rgb)

out = []

for i in range(256):
    # print(i%16,i//16)
    rgbd = rgb[i%16][i//16]
    # print()
    # print()
    out += [",".join([str(i) for i in rgbd])]

with open("turbo2.json","w") as f:
    f.write("{{"+"},\n{".join(out)+"}}")
```
